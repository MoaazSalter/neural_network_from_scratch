# Train the neural network made from scratch on breast_cancer dataset

## Import the data

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target.reshape(-1, 1)
```

```python
X.shape
```

```python
y.shape
```

## Data Preprocessing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=67, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Create the Model

```python
from model.neural_network import NeuralNetwork
from layers.dense import Dense
from optimizers.gradient_descent import Gradient_descent
from activations.sigmoid import Sigmoid
from activations.relu import ReLU
from losses.binary_cross_entropy import Binary_cross_entropy
```

```python
model = NeuralNetwork()
model.add(Dense(30, 16, ReLU()))
model.add(Dense(16, 8, ReLU()))
model.add(Dense(8, 1, Sigmoid()))
```

```python
model.compile(optimizer=Gradient_descent(learning_rate=0.5), loss=Binary_cross_entropy())
```

## Train the model

```python
model.fit(X_train, y_train, epochs=500, verbose=50)
```

## Evaluate the model

```python
import numpy as np
y_pred_train = model.predict(X_train)
train_pred_binary = (y_pred_train > 0.5).astype(int)
train_accuracy = np.mean(train_pred_binary == y_train) * 100
train_accuracy
```

```python
y_pred_test = model.predict(X_test)
test_pred_binary = (y_pred_test > 0.5).astype(int)
test_accuracy = np.mean(test_pred_binary == y_test) * 100
test_accuracy
```

```python
# vonfusion matrix using numpy
true_pos = np.sum((test_pred_binary == 1) & (y_test == 1))
true_neg = np.sum((test_pred_binary == 0) & (y_test == 0))
false_pos = np.sum((test_pred_binary == 1) & (y_test == 0))
false_neg = np.sum((test_pred_binary == 0) & (y_test == 1))

print(f"""
Confusion Matrix (Test Set):
      
                Predicted
                Neg    Pos
Actual  Neg    {true_neg:3d}    {false_pos:3d}
        Pos    {false_neg:3d}    {true_pos:3d}
        """)
```

```python
# calculate precision, recall and the f1 score
precision = true_pos / (true_pos + false_pos) if (true_pos + false_pos) > 0 else 0
recall = true_pos / (true_pos + false_neg) if (true_pos + false_neg) > 0 else 0
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0


print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-Score: {f1:.3f}")
```
